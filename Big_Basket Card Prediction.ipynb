```python
import pandas as pd

# Load the Excel file
dataset = pd.read_excel("BIG_BASKET.xlsx", header=None)

# Initialize the list for transactions
transactions = []

# Loop through the dataset to create the list of transactions
for i in range(0, len(dataset)):
    transactions.append([str(dataset.values[i, j]) for j in range(0, dataset.shape[1])])

# Print the first few transactions to verify
print(transactions[:5])
```

```python
print(dataset)
```

```python
from apyori import apriori
basket = apriori(transactions = transactions , min_support = 0.002,min_confidence = 0.2 , min_lift = 3, min_length = 2 )
```

```python
results = list(basket)
```

```python
print(results)
```

```python
def inspect(results):
    product1 = [tuple(result[2][0][0])[0] for result in results]
    product2 = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidence = [result[2][0][2] for result in results]
    lift  = [result[2][0][3] for result in results]
    return list(zip(product1 , product2 , supports , confidence , lift))
DataFrame_intelligence = pd.DataFrame(inspect(results), columns=['product1','product2','supports','Confidence','lift'])
```
